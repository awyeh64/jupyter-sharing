```python
import pandas as pd
import savReaderWriter as spss
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.plotly as py
import plotly as plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot


file_name = "WELLCOME_APCspend2013_forThinkful.csv"
df = pd.read_csv(file_name, encoding="latin1")

df.head()
```

```python
def journal_title_mapping(j):
    j = str(j)
    if "AMERICAN CHEMICAL SOCIETY" in j or "ACS" in j:
        return "ACS"
    elif "BIOMED CENTRAL" in j or "BMC" in j:
        return "BMC"
    elif "INTERNATIONAL JOURNAL" in j:
        return "INTERNATIONAL JOURNAL"
    else:
        return j
```

```python
# normalizing journal titles
df["Journal_title"]=df['Journal_title'].str.upper()

df["journal_title_clean"] = df.Journal_title.apply(lambda x : journal_title_mapping(x))
```
