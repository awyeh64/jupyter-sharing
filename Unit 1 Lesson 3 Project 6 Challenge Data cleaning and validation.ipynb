```python
def journal_title_mapping(j):
    j = str(j)
    if "AMERICAN CHEMICAL SOCIETY" in j or "ACS" in j:
        return "ACS"
    elif "BIOMED CENTRAL" in j or "BMC" in j:
        return "BMC"
    elif "INTERNATIONAL JOURNAL" in j:
        return "INTERNATIONAL JOURNAL"
    elif "JOURNAL OF NEUROPHYSIOLOGY" in j:
        return "JOURNAL OF NEUROPHYSIOLOGY"
    elif "AMERICAL JOURNAL OF PSYCHIATRY" in j:
        return "AMERICAN JOURNAL OF PSYCHIATRY"
    elif "JOURNAL OF EXPERIMENTAL PSYCHOLOGY" in j:
        return "JOURNAL OF EXPERIMENTAL PSYCHOLOGY"
    elif "ANTIMICROBIAL AGFENTS AND CHEMOTHERAPY" in j:
        return "ANTIMICROBIAL AGENTS AND CHEMOTHERAPY"
    elif "JOURNAL OF VIROLOGY\n" in j:
        return "JOURNAL OF VIROLOGY"
    elif "BIOLOGICIAL CHEMISTRY" in j:
        return "BIOLOGICAL CHEMISTRY"
    elif "BRITISH JOURNAL OF OPHTHALMOLOGY" in j:
        return "BRITISH JOURNAL OF OPTHALMOLOGY"
    elif "JOURNAL OF NEUROLOGY, NEUROSURGERY" in j:
        return "JOURNAL OF NEUROLOGY, NEUROSURGERY & PSYCHIATRY"
    elif "BRITISH JOURNAL " in j:
        return "BRITISH JOURNAL OF THE HISTORY OF SCIENCE"
    elif "AMERICAN JOURNAL OF CLINICAL NUTRITION" in j:
        return "AMERICAN JOURNAL FOR CLINICAL NUTRITION"
    elif "GENETICS " in j:
        return "GENETICS"
    elif "BIOCHIMICA ET BIOOHYSICA ACTA" in j:
        return "BIOCHIMICA ET BIOOHYSICA ACTA"
    elif "JOURNAL OF AFFECTIVE DISORDERS " in j:
        return "JOURNAL OF AFFECTIVE DISORDERS"
    elif "JOURNAL OF BEHAVIOUR THERAPY AND EXPERIMENTAL PSYCHIATRY" in j:
        return "JOURNAL OF BEHAVIOR THERAPY AND EXPERIMENTAL PSYCHIATRY"
    elif "MOLECLUAR & CELLULAR ENDOCRINOLOGY" in j:
        return "MOLECLUAR AND CELLULAR ENDOCRINOLOGY"
    elif "NEUROIMAGE: CLINICAL" in j:
        return "NEUROIMAGE"
    elif "NEUROSCIENCE " in j:
        return "NEUROSCIENCE"
    elif "NEUROPHARMACOLOGY " in j:
        return "NEUROPHARMACOLOGY"
    elif "NEUROPHYSIOLOGIA" in j:
        return "NEUROPSYCHOLOGIA"
    elif "STUDIES IN HISTORY AND PHILOSOPHY OF SCIENCE PART C" in j:
        return "STUDIES IN HISTORY AND PHILOSOPHY OF SCIENCE PART C"
    elif "TRENDS IN NEUROSCIENCE" in j:
        return "TRENDS IN NEUROSCIENCE"
    elif "ACTA CRYSTALLOGRAPHICA" in j:
        return "ACTA D"
    elif "JOURNAL OF VISUALIZED EXPER" in j:
        return "JOURNAL OF VISUALIZED EXPERIMENTS"
    elif "PROCEEDINGS OF THE NATIONAL ACADEMY OF SCIENCES" in j:
        return "PROCEEDINGS OF THE NATIONAL ACADEMY OF SCIENCES"
    elif "CELL DEATH & DISEASE" in j:
        return "CELL DEATH AND DISEASE"
    elif "NEPHROLOGY DIALYSIS TRANSPLANTATION" in j:
        return "NEPHROLOGY DIALYSIS TRANSPLANTATION"
    elif "PLOS " in j:
        return "PLOS"
    elif "PROC NATL ACAD SCI U S A." in j or "PROC NATL ACAD SCI U S A." in j:
        return "PNAS USA"
    elif "BIOCHEMICAL JOURNAL" in j:
        return "BIOCHEMICAL JOURNALS"
    elif "PLOS NEGECTED TROPICAL DISEASES" in j:
        return "PLOS NEGLECTED TROPICAL DISEASES"
    elif "PROCEEDINGS OF ROYAL SOCIETY" in j or "PROCEEDINGS OF THE ROYAL SOCIETY B: BIOLOGICAL SCIENCES" in j:
        return "PROCEEDINGS OF THE ROYAL SOCIETY"
    elif "EUROPEAN CHILD AND ADOLESCENT PSYCHIATTY" in j:
        return "EUROPEAN CHILD AND ADOLESCENT PSYCHIATRY"
    elif "JOURNAL OF AUTISM AND DEVELOPMENT DISORDERS" in j:
        return "JOURNAL OF AUTISM AND DEVELOPMENTAL DISORDERS"
    elif "SOCIAL PSYCHIATRY " in j:
        return "SOCIAL PSYCHIATRY AND PSYCHIATRIC EPIDEMIOLOGY"
    elif "JOURNAL OD CLINICAL ENDOCRINOLOGY" in j:
        return "JOURNAL OF CLINICAL ENDOCRINOLOGY"
    elif "ANNALS OF NEUROLOGY " in j:
        return "ANNALS OF NEUROLOGY"
    elif "BIRTH DEFECTS RESEARCH PART A: CLINICAL AND MOLECULAR TERATOLOGY\xa0 " in j:
        return "BIRTH DEFECTS RESEARCH PART A: CLINICAL AND MOLECULAR TERATOLOGY"
    elif "HUMAN MUTATION" in j:
        return "HUMAN MUTATION"
    elif "MAGNETIC RESONANCE IN MEDICINE" in j:
        return "MAGNETIC RESONANCE IN MEDICINE"
    elif "MOVEMENT DISORDERS" in j:
        return "MOVEMENT DISORDERS"
    elif "DEV WORLD BIOETH." in j or "DEV. WORLD BIOETH" in j:
        return "DEV WORLD BIOETH"
    elif "JOURNAL OF AIDS" in j or "JAIDS" in j:
        return "JOURNAL OF ACQUIRED IMMUNE DEFICIENCY SYNDROMS (JAIDS)"
        
    else:
        return j
    
    
# normalizing journal titles
df["Journal_title"]=df['Journal_title'].str.upper()

df["journal_title_clean"] = df.Journal_title.apply(lambda x : journal_title_mapping(x))
```
