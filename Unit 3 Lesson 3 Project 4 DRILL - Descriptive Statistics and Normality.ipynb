```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal_var = np.random.normal( 50, 10, 100)
plt.hist( normal_var)
mean = np.mean( normal_var)
std = np.std( normal_var)
plt.axvline( x=mean, color="black")
plt.axvline( x=mean + std, color="black")
plt.axvline( x=mean - std, color="black")
plt.show()
```

```python
triangular_var = np.random.triangular( 0, 15, 100, 100)
plt.hist( triangular_var)
mean = np.mean( triangular_var)
std = np.std( triangular_var)
plt.axvline( x=mean, color="black")
plt.axvline( x=mean + std, color="black")
plt.axvline( x=mean - std, color="black")
plt.show()
```
