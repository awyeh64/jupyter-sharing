```python
import pandas as pd
import savReaderWriter as spss
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(style="whitegrid")
file_name = "StudentsPerformance.csv"
df = pd.read_csv(file_name)
df.head()

```

Choose one variable and plot that variable four different ways.

math score

```python
ax = sns.distplot(df["math score"])
```

```python
ax = sns.boxplot(x="math score", data=df)
```

```python
ax = sns.barplot(df['math score'])
```

```python
ax = sns.pointplot(df['math score'])
```

Choose two continuous variables, and plot them three different ways. math score and writing score


```python
ax = sns.scatterplot(x=df['math score'], y=df['writing score'])
```

```python
ax = sns.lineplot(x=df['math score'], y=df['writing score'])
```

```python
ax = sns.scatterplot(x=df['math score'], y=df['parental level of education'])
```

```python
ax = sns.scatterplot(y=df['math score'], x=df['parental level of education'])
plt.xticks(rotation=90)
```

```python
ax = sns.barplot(x=df['parental level of education'], y=df['math score'])
plt.xticks(rotation=90)
```

```python
ax = sns.boxplot(x="parental level of education", y="math score", data=df)
plt.xticks(rotation=90)
```
