```python
import pandas as pd
import savReaderWriter as spss
import matplotlib.pyplot as plt
import numpy as np
file_name = "StudentsPerformance.csv"
df = pd.read_csv(file_name)
df.head()
```

```python
df['avg'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

plt.hist(df['avg'])
plt.ylim(0,300)
plt.show()
```

This graph is to set the averages of each score type: math, reading, and writing.  This is mainly to generalize the scores of each student as outside of outliers, students generally have scores of each category around the same value.  This also helps keeps the graph a bit neater as it combines three values into one.

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title('Some College')
some = df[df['parental level of education'] == 'some college']
plt.boxplot(some['avg'])

plt.subplot(1,3,2)
plt.title('Bachelor\'s Degree')
bachelor = df[df['parental level of education'] == 'bachelor\'s degree']
plt.boxplot(bachelor['avg'])

plt.subplot(1,3,3)
plt.title('Master\'s Degree')
master = df[df['parental level of education'] == 'master\'s degree']
plt.boxplot(master['avg'])
plt.show()
```

Box and whisker graphs were used to here to see the rough percentiles through the medians of the data.  From what we can see, between kids who have parents that have received some college and bachelor's degree, there is not too much variance in their percentiles, though parents with some college have more outliers near the bottom.  On the other hand, kids with parents that have a Master's degree generally have higher median scores, roughly 5% higher than the other two.

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('Standard')
lunch = df[df['lunch'] == 'standard']
plt.plot(lunch['avg'], color='red',label='red')

plt.legend()

plt.subplot(1,2,2)
plt.title('Free/Reduced')
lunch = df[df['lunch'] == 'free/reduced']
plt.plot(lunch['avg'], color='red',label='red')

plt.legend()

plt.show()
```

This was a comparison between kids that had standard lunches versus kids that have free/reduced lunches.  We can conclude that there is no major differences between scores, which is a good sign that the school is providing proper lunches for kids that are in need for food.

```python
plt.figure(figsize=(10,5))

test = df[df['test preparation course'] == 'none']
plt.subplot(1,2,1)
plt.title("None")
plt.hist(test['math score'], alpha=0.5, label='math',color='red')
plt.hist(test['reading score'], alpha=0.5, label='reading', color='blue')
plt.hist(test['writing score'], alpha=0.5, label='writing', color='green')
plt.hist(test['avg'])
plt.xlim([-10,110])
plt.legend()

test = df[df['test preparation course'] == 'completed']
plt.subplot(1,2,2)
plt.title("Completed")
plt.hist(test['math score'], alpha=0.5, label='math',color='red')
plt.hist(test['reading score'], alpha=0.5, label='reading', color='blue')
plt.hist(test['writing score'], alpha=0.5, label='writing', color='green')
plt.xlim([-10,110])
plt.legend()
plt.show()
```

Using the comparisson here, we can tell that the difference between kids that took a test preparation course and the kids that didn't.  On reading and writing scores, kids that have completed the course scored much much higher, while surprisingly the difference wasn't as big for math.
